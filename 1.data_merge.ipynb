```python
import pandas as pd
import os
from sqlalchemy import create_engine
from dotenv import load_dotenv
```

```python
root_folder = 'data-epc'
file_type = '.csv'

data_dump = []
```

```python
def load_file(path):
    if path.endswith('.csv'):
        return pd.read_csv(path)
    return None 
```

```python
for root, dirs, files in os.walk(root_folder): 
    for file in files: 
        if file.lower().endswith(file_type) and file.lower().startswith('certificates'):
            file_path = os.path.join(root,file)
            print(f'Loading: {file_path}')

            df = load_file(file_path)
            df['source_file'] = file_path
            data_dump.append(df)
```

```python
# Combining the datasets as they have the same schema
combined_df = pd.concat(data_dump, ignore_index=True, sort=False)
combined_df.head()
```

# Land Registry Load

```python
def load_folder_csvs(filepath, file_type='.csv', data_dump=[]):
    root_folder = filepath
    
    for root, dirs, files in os.walk(root_folder): 
        for file in files: 
            if file.lower().endswith(file_type):
                file_path = os.path.join(root,file)
                print(f'Loading: {file_path}')

                df = load_file(file_path)
                df['source_file'] = file_path
                data_dump.append(df)
    return data_dump
```

```python
land_registry = load_folder_csvs(filepath='./data-land_registry')
```

```python
def concat_list_of_df(list_item=land_registry):
    try:
        if len(list_item) == 1:
            return list_item[0].copy()
        
        combined = pd.concat(list_item, ignore_index=True, axis=0)
        return combined
    
    except Exception as e:
        print(f'Combining dataframes errored with: {e}')
        
LR_combined = concat_list_of_df()
```

```python
from rapidfuzz import fuzz, process
```

```python
LR_combined['combined_address'] = LR_combined['combined_address'].str.lower()
valid_EPC['combined_address'] = valid_EPC['combined_address'].str.lower()
```

```python
LR_combined['combined_address'] = LR_combined['combined_address'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)

LR_combined['combined_address'] = LR_combined['combined_address'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
valid_EPC['combined_address'] = valid_EPC['combined_address'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)

#
valid_EPC['combined_address'] = valid_EPC['combined_address'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
right_addresses = valid_EPC['combined_address'].unique().tolist()

# Fuzzy Match operation
def fuzzy_merge_row(address, threshold=80):
    match = process.extractOne(address, right_addresses, scorer=fuzz.ratio)
    if match and match[1] >= threshold:
        return match[0], match[1] # Returning the address and score
    return None

LR_combined[['fuzzy_match', 'match_confidence']] = LR_combined['combined_address'].apply(
    lambda x: pd.Series(fuzzy_merge_row(x))
    )

# Now merging 

result = LR_combined.merge(valid_EPC, left_on='fuzzy_match',right_on='combined_address',how='left', indicator=True)
```
