# Level 2 - Feature Engineering

```python
start = qualified_features
start.head()
```

```python
# Visualising each candidate column.
for col in start.columns:
    print(f'Column: {col} -> Sample: {start[col][0]}')
```

```python
# NA Counts
na_count = start.isna().sum().reset_index()
na_count.loc[na_count[0] != 0]
```

```python
# Dropping columns with NA values above threshold.

na_threshold = 15 #percentage
print(f'Before dropping columns with above {na_threshold}% NA values, there are {len(start.columns)}.\n***')

total_rows = start.shape[0]
dropped_columns = []

for col in start.columns:
    col_na = start[col].isna().sum()
    if (col_na/total_rows*100)>na_threshold:
        print(col,total_rows, col_na/total_rows*100)
        start = start.drop(columns=col)
        dropped_columns.append(col)
        
print(f'***\nAfter dropping columns with NA values above {na_threshold}% there are now {len(start.columns)} columns.')
print(f'Columns dropped were: {drop_cols}')
```

```python
# Visualising each candidate column.
count = 0

for col in start.columns:
    count += 1
    print(f'Column: {col} -> Sample: {start[col][0]} -> Dtype: {start[col].dtype} ')

print(f'***Number of columns: {count}')
```

```python
# Converting datetime columns
date_cols = ['deed_date', 'INSPECTION_DATE']

for col in date_cols:
    start[col] = pd.to_datetime(start[col])
```

```python
# Filling NAs - str cols
str_cols = start.select_dtypes(include='object').columns
start[str_cols] = start[str_cols].fillna('Not Available')

num_cols = start.select_dtypes(include=['int64','float64']).columns
for col in num_cols:
    start[col] = start[col].fillna(start[col].mean())
```

```python
# NA Counts
na_count = start.isna().sum().reset_index()
na_count.loc[na_count[0] != 0].sort_values([0], ascending=False)
```

```python
# Datetime NA Values
dt_na_count = start['INSPECTION_DATE'].isna().sum()
print(f'There are {dt_na_count} records with na values for inspection date. \nThis is {dt_na_count/start.shape[0]*100}% of records.')
```

As the inspection date is an indication of "staleness" for the record and this is only under 6%, these records will be removed.

```python
start = start.dropna(subset=['INSPECTION_DATE'])
```
